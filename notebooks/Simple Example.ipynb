```python
from stochastic_approximation import KW, RM, StarSA, RSA, IASA, ACSA
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%load_ext autoreload
```

```python
initial = 30.0
learning_rate = 10.0
n_iter = 100
stepsize = 0.1
a = 0.1
noise_var = 1
lb = -50.0
ub = 50.0
Dx = 100.0
M = 10.0
L = 0.0
response_function = lambda x: -a*x**2 + np.random.normal(scale=noise_var, size=1)
gradient_function = lambda x: -a*x*2 + np.random.normal(scale=noise_var, size=1)

kw = KW(initial, learning_rate, stepsize, response_function, n_iter, lb=lb, ub=ub, sample_path=True)
rm = RM(initial, learning_rate, gradient_function, n_iter, lb=lb, ub=ub, sample_path=True)
star = StarSA(initial, learning_rate, stepsize, response_function, gradient_function, n_iter, lb=lb, ub=ub, sample_path=True)
rsa = RSA(initial, Dx, M, gradient_function, n_iter, lb=lb, ub=ub, sample_path=True)
iasa = IASA(initial, learning_rate, gradient_function, n_iter, lb=lb, ub=ub, sample_path=True)
acsa = ACSA(initial, L, gradient_function, n_iter, lb=lb, ub=ub, sample_path=True)
```

```python
f, ([ax1, ax2, ax3], [ax4, ax5, ax6]) = plt.subplots(2, 3, sharex=True, sharey=True)
plt.subplot(ax1)
plt.plot(kw.optimize())
plt.subplot(ax2)
plt.plot(rm.optimize())
plt.subplot(ax3)
plt.plot(star.optimize())
plt.subplot(ax4)
plt.plot(rsa.optimize())
plt.subplot(ax5)
plt.plot(iasa.optimize())
plt.subplot(ax6)
plt.plot(acsa.optimize())
plt.tight_layout()
```

```python
fig = plt.figure()
plt.plot(kw.optimize())
plt.plot(rm.optimize())
plt.plot(star.optimize(),'--')
plt.plot(rsa.optimize())
plt.plot(iasa.optimize())
plt.plot(acsa.optimize())
plt.legend(['KW', 'RM', 'STAR', 'RSA', 'IASA', 'ACSA'])
```
